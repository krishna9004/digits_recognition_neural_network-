```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Load the MNIST dataset from Keras
# The dataset is split into training and testing sets
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
# Display the number of training samples in the dataset
len(X_train)
```

```python
# Display the number of testing samples in the dataset

len(X_test)
```

```python
# Display the first image from the training dataset using matplotlib
plt.matshow(X_train[0])
```

```python
# Print the label corresponding to the first training image
# This should return '5' indicating that the first image is of the digit 5
y_train[0]
```

```python
# scaling the values between 0 to 1
X_train = X_train / 255
X_test = X_test / 255
```

```python
X_train[0]
```

```python
# Flatten the training and testing images to convert each 28x28 image into a single array of 784 pixels.
# This is necessary for inputting the data into a neural network, as it expects 1D arrays.
X_train_flattened = X_train.reshape(len(X_train), 28 * 28)
X_test_flattened = X_test.reshape(len(X_test), 28 * 28)
```

```python
# Print the shape of the flattened training data
X_train_flattened.shape
```

```python
# Display the flattened pixel values of the first training image
# This will provide a numerical representation of the image data.
X_train_flattened[0]
```

```python
# Initialize a Sequential model from Keras, which allows stacking layers linearly.
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,), activation='sigmoid')
])
# Add a Dense layer with 10 units (one for each digit 0-9).
    # The input shape is set to (784,), which corresponds to the flattened size of the images.
    # The activation function used here is 'sigmoid', which will output values between 0 and 1 for each unit.

# Compile the model with specific parameters
model.compile(optimizer='RMSProp',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
# Evaluate the trained model using the test dataset.
# This function computes the loss and accuracy metrics on the test data (X_test_flattened).
# The output provides a quantitative measure of how well the model performs on unseen data.
model.evaluate(X_test_flattened, y_test)
```

```python
# Use the trained model to predict the labels for the test dataset.
# This function outputs the predicted probabilities for each class (digits 0-9) for all test images.
y_predicted = model.predict(X_test_flattened)

# Display the predicted probabilities for the first test image.
# The output is an array of probabilities corresponding to each digit from 0 to 9.
# Each value indicates the model's confidence that the image represents the corresponding digit.
y_predicted[0]
```

```python
# Visualize the first test image from the dataset using matplotlib.
plt.matshow(X_test[0])
```

```python
# Get the predicted digit (class) for the first test image based on the highest probability.
predicted_digit = np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train_flattened, y_train, epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test,y_test)
```
